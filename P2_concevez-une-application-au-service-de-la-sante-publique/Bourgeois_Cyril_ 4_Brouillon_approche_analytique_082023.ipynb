## Commen√ßons par d√©terminer un seuil de s√©lection pour construire avec des colonnes disposant d'un maximum de donn√©es :

---

```python
def null_factor(df, tx_threshold=50):
  null_rate = ((df.isnull().sum() / df.shape[0])*100).sort_values(ascending=False).reset_index()
  null_rate.columns = ['Variable','Taux_de_Null']
  high_null_rate = null_rate[null_rate.Taux_de_Null >= tx_threshold]
  return high_null_rate
```

```python
full_null_rate = null_factor(df, 100)
full_null_rate
```

```python
filling_features = null_factor(df, 0)
filling_features["Taux_de_Null"] = 100-filling_features["Taux_de_Null"]
filling_features = filling_features.sort_values("Taux_de_Null", ascending=False) 

#Seuil de suppression
sup_threshold = 20

fig = plt.figure(figsize=(20, 35))

font_title = {'family': 'serif',
              'color':  '#394f18', #'#114b98',
              'weight': 'bold',
              'size': 18,
             }
sequential_colors = sns.color_palette("YlGnBu")
#sequential_colors = sns.color_palette("Set3")
sns.barplot(x="Taux_de_Null", y="Variable", data=filling_features, palette=sequential_colors)
#Seuil pour suppression des varaibles
plt.axvline(x=sup_threshold, linewidth=2, color = 'r')
plt.text(sup_threshold+2, 65, 'Seuil de suppression des variables', fontsize = 16, color = 'r')

plt.title("Taux de remplissage des variables dans le jeu de donn√©es (%)", fontdict=font_title)
plt.xlabel("Taux de remplissage (%)")
plt.show()
```

```python
#Liste des variables √† conserver
features_to_conserve = list(filling_features.loc[filling_features['Taux_de_Null']>=sup_threshold, 'Variable'].values)
#Liste des variables supprim√©es
deleted_features = list(filling_features.loc[filling_features['Taux_de_Null']<sup_threshold, 'Variable'].values)
```

```python
# S√©lectionner les lignes en duplicata sur la colonne "code"
duplicate_rows = datas[datas.duplicated(subset=["code"], keep=False)]

# Cr√©er un dictionnaire pour stocker les informations les plus compl√®tes pour chaque code
code_info = {}

# Parcourir les lignes en duplicata
for idx, row in duplicate_rows.iterrows():
    code = row["code"]
    if code not in code_info:
        code_info[code] = row
    else:
        existing_row = code_info[code]
        # Remplacer les valeurs manquantes de existing_row avec les valeurs de row
        for col in datas.columns:
            if pd.isnull(existing_row[col]) and not pd.isnull(row[col]):
                existing_row[col] = row[col]

# Convertir le dictionnaire en DataFrame
unified_data = pd.DataFrame(list(code_info.values()))

# Afficher le DataFrame r√©sultant
unified_data
```
